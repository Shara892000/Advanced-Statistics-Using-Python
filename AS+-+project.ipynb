Salary is hypothesized to depend on educational qualification and occupation. To understand the dependency, the salaries of 40 individuals [SalaryData.csvView in a new window] are collected and each personâ€™s educational qualification and occupation are noted. Educational qualification is at three levels, High school graduate, Bachelor's, and Doctorate. Occupation is at four levels, Administrative and clerical, Sales, Professional or specialty, and Executive or managerial. A different number of observations are in each level of education â€“ occupation combination.

## 1.State the null and the alternate hypothesis for conducting one-way ANOVA for both Education and Occupation individually.

education_null_hypothesis = "The mean salary is the same across all education levels (High school graduate, Bachelor's, Doctorate)."
education_alternate_hypothesis = "The mean salary is different for at least one level of education."

occupation_null_hypothesis = "The mean salary is the same across all occupation levels (Administrative and clerical, Sales, Professional or specialty, Executive or managerial)."
occupation_alternate_hypothesis = "The mean salary is different for at least one level of occupation."



```python
print(f"Null hypothesis for Education: {education_null_hypothesis}")
print(f"Alternate hypothesis for Education: {education_alternate_hypothesis}")
print(f"Null hypothesis for Occupation: {occupation_null_hypothesis}")
print(f"Alternate hypothesis for Occupation: {occupation_alternate_hypothesis}")
```

## 2. Perform a one-way ANOVA on Salary with respect to Education. State whether the null hypothesis is accepted or rejected based on the ANOVA results.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('Salary ~ Education', data=ano).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
alpha = 0.05 
p_value = anova_table['PR(>F)'][0]

if p_value < alpha:
    print(f"Reject the null hypothesis. There is evidence that Salary depends on Education.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence that Salary depends on Education.")
```

## 3. Perform a one-way ANOVA on Salary with respect to Occupation. State whether the null hypothesis is accepted or rejected based on the ANOVA results.

```python

import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('Salary ~ Occupation', data=ano).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
alpha = 0.05  # Set the significance level
p_value = anova_table['PR(>F)'][0]

if p_value < alpha:
    print(f"Reject the null hypothesis. There is evidence that Salary depends on Occupation.")
else:
    print("Fail to reject the null hypothesis. There is no significant evidence that Salary depends on Occupation.")
```

## 5. Perform a two-way ANOVA based on Salary with respect to both Education and Occupation (along with their interaction Education*Occupation). State the null and alternative hypotheses and state your results. How will you interpret this result?

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

model = ols('Salary ~ Education * Occupation', data=ano).fit()
anova_table = anova_lm(model, typ=2)

print(anova_table)


alpha = 0.05 
p_value = anova_table['PR(>F)'][0]

if p_value < alpha:
    print("Reject the null hypothesis. There is evidence of a significant difference in mean Salary.")
else:
    print("Fail to reject the null hypothesis. No significant evidence of a difference in mean Salary.")


interaction_p_value = anova_table['PR(>F)']['Education:Occupation']
if interaction_p_value < alpha:
    print("The interaction effect between Education and Occupation is significant.")
else:
    print("The interaction effect between Education and Occupation is not significant.")
```
